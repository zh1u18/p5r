```python
import pandas as pd
from urllib.parse import quote

url = r'https://wiki.biligame.com/persona/%E4%BA%BA%E6%A0%BC%E9%9D%A2%E5%85%B7'
tables = pd.read_html(url)
df_overall = tables[1].iloc[:, 1:]
df_overall.to_csv('p5r_人格面具.csv')
df_overall
```

```python
import time

persona_names = df_overall.iloc[:, 0].to_list()
quote_names = [quote(x) for x in persona_names]
data_dict = {}
for idx, quote_name in enumerate(quote_names):
    try:
        url = f'https://wiki.biligame.com/persona/P5R/{quote_name}'
        tables = pd.read_html(url)
        data_dict[persona_names[idx]] = tables
        time.sleep(0.03)
    except Exception as e:
        print(e)
        print(quote_name)
        data_dict[persona_names[idx]] = []
        
num_tables = [len(tables) for tables in data_dict.values()]
```

```python
import pickle

with open('p5r_data.pkl', 'wb') as file:
    pickle.dump(data_dict, file)
```
