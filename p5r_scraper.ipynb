```python
import pandas as pd
from urllib.parse import quote

url = r'https://wiki.biligame.com/persona/%E4%BA%BA%E6%A0%BC%E9%9D%A2%E5%85%B7'
tables = pd.read_html(url)
df_overall = tables[1].iloc[:, 1:]
df_overall.to_csv('p5r_人格面具.csv')
df_overall
```

```python
url = r'https://wiki.biligame.com/persona/P5R%E9%81%93%E5%85%B7%E5%9B%BE%E9%89%B4'
tables = pd.read_html(url)
df_overall = tables[2]
df_overall.to_csv('p5r_道具图鉴.csv')
df_overall
```

```python
import time

persona_names = df_overall.iloc[:, 0].to_list()
quote_names = [quote(x) for x in persona_names]
data_dict = {}
for idx, quote_name in enumerate(quote_names):
    try:
        url = f'https://wiki.biligame.com/persona/P5R/{quote_name}'
        tables = pd.read_html(url)
        data_dict[persona_names[idx]] = tables
        time.sleep(0.03)
    except Exception as e:
        print(e)
        print(quote_name)
        data_dict[persona_names[idx]] = []
        
num_tables = [len(tables) for tables in data_dict.values()]
```

```python
import pickle

with open('p5r_data.pkl', 'wb') as file:
    pickle.dump(data_dict, file)
```

```python
dfs = []
columns = ['价格', '种族', 'Lv', 'Name', '种族', 'Lv', 'Name', 'Result']

for k, v in data_dict.items():
    if k in skip_persona + skip_persona2:
        continue
    else:
        df_temp = v[-1].iloc[4:, :]
        df_temp.loc[:, 'result'] = k
        dfs.append(df_temp)
df_full = pd.concat(dfs)
df_full.columns = columns
df_full['价格'] = df_full['价格'].astype(float)
df_full.to_csv('p5r_合成总表.csv', index=False)
```
